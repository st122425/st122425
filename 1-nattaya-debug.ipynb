Insertion Sort

```python
import numpy as np
def insertsion_sort(sl):
    #sl = [5, 6, 11, 10, 11, 100, 33, 55, 44]
    num_list = len(sl)

    for j in range(1, num_list):
        key = sl[j]
        i = j-1
        while i >= 0 and sl[i] > key:
            sl[i+1] = sl[i]
            i = i-1
        sl[i+1] = key
        j = j+1

    print(f"Ordered numbers: {sl}")

insertsion_sort([1, 9, 10, 100, 33, 22, 1, 3, 4])
    
```
