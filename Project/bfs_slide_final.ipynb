```python
def convert(matrix):
    list = []
    for i in range(3):
        for j in range(3):
            list.append(matrix[i][j])
    return list
```

```python
def check(cur, goal):
    done=True
    for i in range(len(cur)):
        if(cur[i] != goal[i]):
            done = False
    return done
```

```python
def possible_st(cur, goal):
    posible_st = []
    pos = cur.index(0)
    if(pos == 0):
        go = [2, 4]
    elif(pos == 1):
        go = [1, 2, 4]
    elif(pos == 2):
        go = [1, 4]
    elif(pos == 3):
        go = [2, 3, 4]
    elif(pos == 4):
        go = [1, 2, 3, 4]
    elif(pos == 5):
        go = [1, 3, 4]
    elif(pos == 6):
        go = [2, 3]
    elif(pos == 7):
        go = [1, 2, 3]
    else:
        go = [1, 3]
  
    for g in go:
        #left
        if(g == 1):
            n_post = cur[:]
            n_post[pos-1], n_post[pos] = n_post[pos], n_post[pos-1]
            posible_st.append(n_post)
        #right
        if(g == 2):
            n_post = cur[:]
            n_post[pos+1], n_post[pos] = n_post[pos], n_post[pos+1]
            posible_st.append(n_post)
        #up
        if(g == 3):
            n_post = cur[:]
            n_post[pos-3], n_post[pos] = n_post[pos], n_post[pos-3]
            posible_st.append(n_post)

        #down
        if(g == 4):
            n_post = cur[:]
            n_post[pos+3], n_post[pos] = n_post[pos], n_post[pos+3]
            posible_st.append(n_post)
            
    return posible_st
```

```python
def print_matrix(list):
    for (index, value) in enumerate(list):
        print(' %s ' % value, end=' ') 
        if index in [x for x in range(2, 9, 3)]:
            print() 
    print() 
```

```python

def solve(s, g):
    possible_move = deque()
    moved = {repr(s):None}
    cur = [ele for ele in s]
    while not check(cur, g):
        pos_move = possible_st(cur, g)
        for next_st in pos_move:
            if repr(next_st) not in moved:
                possible_move.appendleft(next_st)
                moved[repr(next_st)] = cur
        try:
            cur = possible_move.pop()
        except IndexError:
            return 0
    solve = []
    solve.append(cur)
    while not check (cur, s):
        cur = moved[repr(cur)]
        solve.append(cur)
    reversed2 = solve[::-1]
    return reversed2
```

```python
from collections import deque
from time import process_time

start = [[2, 8, 3],
         [1, 6, 4],
         [7, 0, 5]]

goal = [[1, 2, 3],
        [8, 0, 4],
        [7, 6, 5]]

count = 0

print("----Start stage----")
start_list = convert(start)
print(start_list)
print_matrix(start_list)


print("----Goal stage----")
goal_list = convert(goal)
print_matrix(goal_list)

print("----Start solving with bfs----")
t1_start = process_time()
solution = solve(start_list,goal_list)
t1_stop = process_time()
if solution == 0:
    print("Cannot find the optimal solution")
else:
    for ele in solution:
        print("step:", count)
        print_matrix(ele)
        count = count + 1
print("Process time:", t1_stop-t1_start)
```
